## 서울 미세먼지 상황

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
showme=urlopen("http://cleanair.seoul.go.kr/air_pollution.htm?method=average")
showme
```

```python
bs=BeautifulSoup(showme,"html.parser")
bs
```

```python
p=bs.find('map',{'id':"chart2Map"})
print(p)
```

```python
for area in p.find_all('area'):
    print(area.get('alt'))
```

```python
with open('crawling_2.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['지역 - 미세먼지'])
    
    for area in p.find_all('area'):
        content=area.get('alt')
        writer.writerow([content])
```
