```python
# _*_ coding: UTF-8 _*_
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen("https://music.naver.com/home/index.nhn")
html
```

```python
bs=BeautifulSoup(html,"html.parser")
bs
```

```python
tr1=bs.find("tr",{"class":"_tracklist_move _track_dsc list1 on"})
print(tr1)
for n in range(2,11):
    tr2=bs.find("tr",{"class":"_tracklist_move _track_dsc list"+str(n)+" "})
    print(tr2)
```

```python
title1=tr1.find("td",{"class":"name"}).get_text()
print(title1)
for n in range(2,11):
    tr2=bs.find("tr",{"class":"_tracklist_move _track_dsc list"+str(n)+" "})
    title2=tr2.find("td",{"class":"name"}).get_text()
    print(title2)
```

```python
artist1=tr1.find("td",{"class":"_artist artist"}).get_text()
print(artist1)
for n in range(2,11):
    tr2=bs.find("tr",{"class":"_tracklist_move _track_dsc list"+str(n)+" "})
    artist2=tr2.find("td",{"class":"_artist artist"}).get_text()
    print(artist2)
```

```python
album1=tr1.find("td",{"class":"album"}).get_text()
print(album1)
for n in range(2,11):
    tr2=bs.find("tr",{"class":"_tracklist_move _track_dsc list"+str(n)+" "})
    album2=tr2.find("td",{"class":"album"}).get_text()
    print(album2)
```
