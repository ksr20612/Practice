## 전국 현재 대기오염 정보

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html=urlopen('http://www.kweather.co.kr/main/main.html')
html
```

```python
bs=BeautifulSoup(html,'html.parser')
bs
```

```python
div=bs.find("div",{"class":"local"})
div
```

```python
for ul in div.find_all('ul'):
    area=ul.span.get_text()
    degree=ul.img.get('alt')
    print(area,degree)
```

```python
with open('crawling_practice5(3.26).csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['지역', '상태'])
    
    for ul in div.find_all('ul'):
        area=ul.span.get_text()
        degree=ul.img.get('alt')
        writer.writerow([area,degree])
```
