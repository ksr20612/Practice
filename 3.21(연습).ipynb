### 네이버 뮤직 차트 1~10위 노래 제목과 가수이름 따오기

```python
# _*_ coding: UTF-8 _*_
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen("https://music.naver.com")
html
```

```python
bs = BeautifulSoup(html, "html.parser")
print(bs)
```

```python
range(1,10)
```

```python
bs.head.find_all('meta')
```

1. 랭킹들 사이에서 차이점 찾기

```python
for n in range(1,2):
    tr1=bs.find("tr",{"class":"_tracklist_move _track_dsc list"+str(n)+" on"})
    print(tr1)
for m in range(2,11):
    tr2=bs.find("tr",{"class":"_tracklist_move _track_dsc list"+str(m)+" "})
    print(tr2)

tr1+':'+tr2
```

```python
for m in range(2,11):
    tr2=bs.find("tr",{"class":"_tracklist_move _track_dsc list"+str(m)+" "})
    print(tr2)
```

2. 차이가 있는 태그에서 앨범명 뽑기

```python
for n in range(1,2):
    tr1=bs.find("tr",{"class":"_tracklist_move _track_dsc list"+str(n)+" on"})
    album1=tr1.find("td",{"class":"album"}).get_text()
    print(album1)
for m in range(2,11):
    tr2=bs.find("tr",{"class":"_tracklist_move _track_dsc list"+str(m)+" "})
    album2=tr2.find("td",{"class":"album"}).get_text()
    print(album2)
```

3. 가수 뽑기

```python
singer=tr1.find("a",{"class":"_artist"}).get_text()
print(singer)
```

4. 노래명 뽑기

```python
song=tr1.find("td",{"class":"name"}).get_text()
print(song)
```
