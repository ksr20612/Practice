## 다음 실시간 검색어

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html=urlopen('https://www.daum.net/')
html
```

```python
bs=BeautifulSoup(html,'html.parser')
bs
```

```python
ol=bs.find("ol",{"class":"list_hotissue issue_row"})
ol
```

```python
for n in range(1,11):
    span=ol.find("span",{"class":"num_pctop rank"+str(n)})
    rank=span.get_text()
    content=span.next_sibling.next_sibling.get_text()
    print(rank, content)
```

```python
with open('crawling_practice3.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['순위', '검색어'])
    
    for n in range(1,11):
        span=ol.find("span",{"class":"num_pctop rank"+str(n)})
        rank=span.get_text()
        content=span.next_sibling.next_sibling.get_text()
        writer.writerow([rank, content])
```
